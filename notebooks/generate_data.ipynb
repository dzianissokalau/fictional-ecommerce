```python
# select users at random
# TO DO:
# need to update probability of visit at the session end
# create more users but set probability very low, so the number of new visitors will not exceed a treshold
total_n_users = 2000
user_ids = np.arange(1, total_n_users+1, 1)

initial_prob = 0.05
visit_probs = np.linspace(initial_prob, initial_prob, num=total_n_users)

visitors_index = [np.random.default_rng(seed=abs(hash(datetime.now()))).choice(a=[True,False], p=[x, 1-x]) for x in visit_probs]
visitors = user_ids[visitors_index]
```

```python
start_date = datetime.now() - timedelta(days=1)


for visitor in visitors:
    print('\nUSER: {}'.format(visitor))
    
    if visitor not in users.keys():
        users[int(visitor)] = User(name='{} user'.format(visitor), user_id=int(visitor))
    
    timestamp = start_date + timedelta(seconds=int(visitor)*5)
    
    rand = np.random.default_rng(seed=visitor)
    session(user_id=int(visitor), 
            timestamp=timestamp, 
            platform='ios', 
            country='US', 
            events=events, 
            satisfaction_impact=satisfaction_impact)
```
